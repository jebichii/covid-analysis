```python
# Add death rate column
subset['death_rate'] = subset['total_deaths'] / subset['total_cases']

plt.figure(figsize=(12, 6))
sns.lineplot(data=subset, x='date', y='death_rate', hue='location')
plt.title('COVID-19 Death Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Death Rate')
plt.legend(title='Country')
plt.tight_layout()
plt.show()
```

```python
# Get the latest total cases per country
latest = df.dropna(subset=['total_cases'])
latest = latest.groupby('continent')['total_cases'].max().sort_values(ascending=False).head(10)

# Plot bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=latest.values, y=latest.index, palette='coolwarm')
plt.title('Top 10 Countries by Total COVID-19 Cases')
plt.xlabel('Total Cases')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```
